```python
import matplotlib.pyplot as plt
import numpy as np
import csv
%matplotlib inline
```

# Plot the output of FC.xx (FCF Spectrum)
We took the spectrum from the outfile into a csv file so that we could plot it nicely
In order to get the Duschinsky matrix and the delta vector correctly, we inserted print statements in the FC.xx code and got it out that way

```python
#The V3.spec.out content is in the csv file V3_spectrum
xaxis = [0.0 for i in range(0,2000+300)]
yaxis = [0.0 for i in range(0,2000+300)]
ctr = 0

with open('V3_spectrum.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:
        xaxis[ctr] = float(row[0])
        yaxis[ctr] = float(row[1])
        ctr += 1
    
```

```python
fig, ax = plt.subplots()

for i in range(0,2000+300):
    ax.scatter(xaxis[i],yaxis[i],c='b')

plt.xlabel("Position ($cm^{-1}$)")
plt.ylabel("FCF")
plt.rcParams["figure.figsize"] = (30,10)
plt.show()
```
